### Extracting City Names From Shop Names

```python
def cityextractor():
    
    lists = []
    count = 0
    for i in Shop_Names['shop_name']:
        temp = i
        start_pt = temp.find("\"")
        end_pt = temp.find("\"", start_pt + 1) 
        quote = temp[start_pt + 1: end_pt]

        if(len(i)-1 != len(quote)):
            lists.append(quote)
        else:
            lists.append('N/A')
            
    # labeling city names
    cities = []
    indexes = []

    for i, value in enumerate(lists):

        if value not in cities:
            cities.append(value)
            indexes.append(len(cities)-1)
        else:
            for x, v in enumerate(cities):
                if v == value:
                    indexes.append(x)
    
    return indexes     
```

### Removing Outliers

```python
def outliers(feature):
    
    sum_value = 0
    for i in feature:
        sum_value += i
    
    mean = sum_value / len(feature)
    
    std = 0
    value = 0
    for i in feature:
        value += (i - mean) ** 2
        
    std = (value / len(feature)) ** 0.5
    
    upperlimit = mean + 3* std
    lowerlimit = mean - 3* std
    
    outlier, index = [], []
    for i, value in enumerate(feature):
        if value >= upperlimit or value <= lowerlimit:
            index.append(i)
            outlier.append(value)
            
    return index, outlier
```

```python
from itertools import product
def gridder(sale_train):
    grid = []

    for block_num in sale_train['date_block_num'].unique():

        cur_shops = sale_train[sale_train['date_block_num']==block_num]['shop_id'].unique()

        cur_items = sale_train[sale_train['date_block_num']==block_num]['item_id'].unique()

        grid.append(np.array(list(product(*[cur_shops, cur_items, [block_num]])),dtype='int32'))



    #turn the grid into pandas dataframe

    index_cols = ['shop_id', 'item_id', 'date_block_num']

    grid = pd.DataFrame(np.vstack(grid), columns = index_cols,dtype=np.int32)
    return grid
```

```python
def scale_features(X):
        
        df = pd.DataFrame(X)
        
        mini , maxi = [] , []
        for i in df.columns:

            mini.append(min(df[i]))
            maxi.append(max(df[i]))
            
        xmin , xmax = np.array(mini) , np.array(maxi)
        X = (X - xmin) / (xmax - xmin)

        return X
```

```python
import numpy as np

class LassoRegression:
    def __init__(self, alpha=1.0, max_iter=1000, tol=1e-6):
        self.alpha = alpha
        self.max_iter = max_iter
        self.tol = tol
        self.coef_ = None
        self.intercept_ = None
        self.history_ = None
    
    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.coef_ = np.zeros(n_features)
        self.intercept_ = 0
        self.history_ = []
        
        for i in range(self.max_iter):
            prev_coef = np.copy(self.coef_)
            for j in range(n_features):
                residual = y - np.dot(X, self.coef_) - self.intercept_
                self.coef_[j] += 2 * X[:, j].dot(residual) / n_samples
                self.coef_[j] = np.sign(self.coef_[j]) * max(0, np.abs(self.coef_[j]) - self.alpha)
            self.intercept_ = np.mean(y - np.dot(X, self.coef_))
            
            self.history_.append(np.sum(np.abs(prev_coef - self.coef_)))
            if np.sum(np.abs(prev_coef - self.coef_)) < self.tol:
                break
    
    def predict(self, X):
        return np.dot(X, self.coef_) + self.intercept_
```

#### Self Made - Linear Regression

```python
class LinearRegression:
    
    # Constructor
    def __init__(self, alpha, epochs):  
        
        self.alpha = alpha
        self.thetas = []
        self.epochs = epochs
        
    # Hypothesis Function
    def hypothesis(self, X , thetas):
        
        return np.dot(X , thetas)
    
    # Loss Function
    def loss_function(self, actual , predicted):
        
        return np.mean((actual - predicted)**2)
    
    # Derivative Of Loss Function
    def derivative_loss(self, X, Y, thetas):
    
        h = self.hypothesis(X, thetas)
        difference = Y - h
        r = np.dot(difference.T, X)
        r = -2*r / float(X.shape[0])

        return r
    
    # Gradient Descent Function
    def gradient_descent(self, o_thetas, g_thetas, alpha):
        
        return (o_thetas - alpha * g_thetas)
    
    # Performing All Calculations For Thetas
    def Calculations(self, X, Y, thetas, alpha):

        predictions = self.hypothesis(X , thetas)
        MSE = self.loss_function(Y , predictions)

        g_thetas = self.derivative_loss(X, Y, thetas).T
        new_thetas = self.gradient_descent(thetas, g_thetas, alpha)

        return new_thetas, MSE
    
    # Train Function
    def train(self, X, Y):
     
        one = np.ones((X.shape[0], 1))
        X = np.append(X, one, axis = 1)
        Y = Y.reshape(-1,1)
        
        self.thetas = np.zeros((X.shape[1],1))       
        MSE = 0
        for i in range(self.epochs):
            self.thetas, MSE = self.Calculations(X, Y, self.thetas, self.alpha) 
            if i % 1000 == 0:
                print(i)
        
    # Test Function
    def test(self, X):
        
        one = np.ones((X.shape[0], 1))
        X = np.append(X, one, axis = 1)
        
        return self.hypothesis(X, self.thetas)
```

#### Self Made - K Nearest Neighbor

```python
class KNearestNeighbor:

    # Constructor
    def __init__(self, k):  
        
        self.k = k
        self.X_train = []    
        self.Y_train = [] 
   
    # Computing Distances Of Train Set
    def compute_distances_no_loops(self, X):

        num_test = X.shape[0]
        num_train = self.X_train.shape[0]
        
        dists = np.zeros((num_test, num_train)) 
        dists = np.sqrt((X**2).sum(axis=1)[:, np.newaxis] + (self.X_train**2).sum(axis=1) - 2 * X.dot(self.X_train.T))
        return dists

    # Training Function
    def train(self, X, Y):
        
        nexamples , nfeatures = X.shape
            
        self.X_train = X    
        self.Y_train = Y 
    
    def test(self, X):
        
        num_test = X.shape[0]
        
        y_pred = np.zeros(num_test, dtype = self.Y_train.dtype)
        
        compute_distance = self.compute_distances_no_loops

        dists=compute_distance(X)
        for i , value in enumerate(dists): # picking each example
            
            sorted_array = np.argsort(value) # sorting

            nearest_neighbors = [] 
            for j in range(self.k): # picking neighbour one by one
                
                index = sorted_array[j]
                nearest_neighbors.append(self.Y_train[index])

            y_pred[i] = np.mean(nearest_neighbors)
            
        return y_pred
```
